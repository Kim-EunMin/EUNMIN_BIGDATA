```python
# 1. null 제거
NA = flight['changed_time'].isnull()
NA_INDEX = NA[NA==True].index
NA_INDEX

flight.drop(NA_INDEX,inplace=True)
```

```python
# 2. 형식 바꾸기 ( 시간 차를 구하기 위해서)
flight['scheduled_time2'] = flight.scheduled_time.apply(lambda x: str(x)[:-2]+'-'+ str(x)[-2:])
```

```python
import math
#float->int형으로 교환
flight['changed_time2'] = flight.changed_time.apply(lambda x: math.trunc(x))
```

```python
flight.changed_time2 = flight.changed_time2.apply(lambda x: str(x)[:-2]+'-'+str(x)[-2:])
```

```python
flight.head()
```

```python
# 시간 차이 구하기
import datetime
flight.scheduled_time2=flight.scheduled_time2.apply(lambda x: datetime.datetime.strptime(x, '%H-%M'))
flight.changed_time2 = flight.changed_time2.apply(lambda x:datetime.datetime.strptime(x, '%H-%M'))
```

```python
flight['diff'] = flight.changed_time2 - flight.scheduled_time2
```

```python
flight['diff'] = flight['diff'].apply(lambda x: x.total_seconds())//60
```

```python
flight['diff'] = flight['diff'].apply(lambda x: math.trunc(x))
```

```python
del flight['scheduled_time2']
del flight['changed_time2']
```

```python
flight.head()
```

```python
# 1~5분, 5~10분 단위로, 10~20분 단위로, 20~30분
# 0일때는 (delay가 없다 ) : delay_section 이 0
# 1~5분 : delay_section 이 1이다
# 5~10분 : delay_section 이 2이다
# 10~20분 : delay_Section 이 3이다
# 20~30분 : delay_section 이 4이다.
# 결항 : 5번 그룹
# 30분 이상은 6번그룹
```

```python
flight['delay_section'] = 6 # 변수를 미리 생성
```

```python
# delay 없다. => 0번집단
flight.loc[(flight['diff']<=0), ['delay_section']] = 0
```

```python
#delay 가 1분부터 5분까지 => 1번집단
flight.loc[(flight['diff']>0)& (flight['diff']<=5), ['delay_section']] = 1
```

```python
#delay 가 6분부터 10분까지 =>2번집단

flight.loc[(flight['diff']>5)& (flight['diff']<=10), ['delay_section']] = 2
```

```python
#delay 가 11분부터 20분까지 =>3번집단
flight.loc[(flight['diff']>10)& (flight['diff']<=20), ['delay_section']] = 3
```

```python
#delay 가 21분부터 30분까지 =>4번집단
flight.loc[(flight['diff']>20)& (flight['diff']<=30), ['delay_section']] = 4
```

```python
# 결향인 것은 5번 집단으로 한다.
flight.loc[(flight['flight_status_kor']=='결항'), ['delay_section']] = 5
```

```python
# 일별 지연횟수 확인하고자한다
```

```python
#flight['create_date'][0].month
```

```python
flight['day'] = flight['create_date'].apply(lambda x: str(x.month) + '-' + str(x.day))
```

```python
flight.head()
```

```python
# 일별 전체 비행 횟수 도수 분포표
flight_count = flight.groupby(by=['day'], as_index=False).count()
flight_count = flight_count.iloc[:,:2]
flight_count.set_index(flight_count['day'],inplace=True)
flight_count.drop('day',axis=1 , inplace=True)
```

```python
# 도착 지연된 비행 횟수를 도수분포표로 표현한다. ( diff가 양수인게 도착이 예상보다 지연된 것이다 ) 
arrive_delay = flight[flight['diff']>0]
arrive_delay_count = arrive_delay.groupby(by=['day'], as_index=False).count()
arrive_delay_count = arrive_delay_count.iloc[:,:2]
arrive_delay_count.set_index(arrive_delay_count['day'],inplace=True)
arrive_delay_count.drop('day',axis=1 , inplace=True)
```

```python
# 출발 지연된 비행 횟수를 도수분포표로 표한한다 ( 지연 Y 라고 표시된 것이 출발 지연된 것이다 ) 
depart_delay = flight[flight['delay_yn']=="Y"]
depart_delay_count = depart_delay.groupby(by=['day'], as_index=False).count()
depart_delay_count = depart_delay_count.iloc[:,:2]
depart_delay_count.set_index(depart_delay_count['day'],inplace=True)
depart_delay_count.drop('day',axis=1 , inplace=True)
#depart_delay_count.head()
```

```python
df = pd.merge(flight_count,arrive_delay_count, left_index=True, right_index=True, how='left' )
df = pd.merge(df, depart_delay_count,left_index=True, right_index=True, how='left' )
```

```python
df.columns = ["total_flight_num", "arrive_delay_count", "depart_delay_count"] 
#전체 비행횟수, 도착지연횟수, 출발지연횟수 순으로 표시
```
